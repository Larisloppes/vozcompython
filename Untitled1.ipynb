```python
engine = pyttsx3.init()
```

```python
voices = engine.getProperty('voices')
for voice in voices:
    print("Voz:")
    print(" - Nome: %s" % voice.name)
    if voice.languages:
        print(" - Idioma: %s" % voice.languages[0])
    else:
        print(" - Idioma: Desconhecido")
    print(" - Gênero: %s" % voice.gender if voice.gender else "Desconhecido")
```

```python
def selecionar_voz_feminina_pt_BR():
    engine = pyttsx3.init()
    voices = engine.getProperty('voices')
    for voice in voices:
        if voice.languages and 'brazil' in voice.languages[0].lower() and 'female' in voice.name.lower():
            engine.setProperty('voice', voice.id)
            break
    return engine
```

```python
def sintetizar_audio(engine, texto, nome_arquivo):
    engine.save_to_file(texto, nome_arquivo)
    engine.runAndWait()
```
